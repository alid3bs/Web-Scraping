```python
""""
importing selenium package for web scrabing and pandas for data preprosessing and exploring
"""""
from selenium import webdriver
import pandas as pd

driver = webdriver.Firefox()
```

```python
""""
going to www.premierleague.com to make web scrabing
"""""
driver.get('https://www.premierleague.com/stats/top/players/goals?se=-1')
```

```python
""""
Extracting player names from the website
"""""
player_name=driver.find_elements_by_xpath('//a[@class="playerName"]')

player_name_list=[]

for player in player_name:
    player_name_list.append(player.text)
```

```python
player_name_list[:5]
```

```python
""""
Extracting player ranks from the website
"""""

player_rank=driver.find_elements_by_xpath('//td[@class="rank"]')

player_rank_list=[]

for rank in player_rank:
    player_rank_list.append(rank.text)
```

```python
player_rank_list[:5] 
```

```python
""""
Extracting player names from the website
"""""


player_team=driver.find_elements_by_xpath('//tr/td[@class="hide-s"][position()  = 1  ]')

player_team_list=[]

for team in player_team:
        player_team_list.append(team.text)
    
```

```python
player_team_list
```

```python
""""
Extracting player nationality from the website
"""""
player_Nationality=driver.find_elements_by_xpath('//span[@class="playerCountry"]') 

player_Nationality_list=[]

for country in player_Nationality:
    player_Nationality_list.append(country.text) 
    
```

```python
player_Nationality_list
```

```python
""""
Extracting player goals from the website
"""""

player_goals=driver.find_elements_by_xpath('//td[@class="mainStat text-centre"]') 

player_goals_list=[]

for goal in player_goals:
    player_goals_list.append(goal.text) 
```

```python
player_goals_list
```

```python
""""
Here I will make my own DataFrame (new data) using the data that be extracted from the website
"""""

new_data=pd.DataFrame(list(zip(player_name_list,player_rank_list,player_team_list,
                               player_Nationality_list,player_goals_list)),
                      columns=['Name','player_rank','player_club','player_Nationality','player_goals'])
new_data
```

```python
df = pd.read_json('players_stats.json')
df
```

```python
""""
Merging the 2 datasets
"""""
df_new=pd.merge(df,new_data,how="outer",on=['Name'])
df_new
```

```python
""""
after merging the two datasets, I will fill manualy the null values for Mohammed Salah
"""""

df_new.at[3,'player_Nationality']='Egyptian'
df_new.at[3,'player_club']='Liverpool'
df_new.at[3,'player_goals']=110
df_new.at[3,'player_rank']=21


df_new.head()
```

```python
""""
here I will encode the player_Nationality to confert it from categorical to numerical featrure, so I can complete
my data preprocessing without problems
"""""
Encoded_df = pd.get_dummies(df_new, columns = ['player_Nationality'])
Encoded_df.head()
```

```python
""""
- here I will drop the player_club column becuse it has many null values
- Then I will drop the Name column also to be able to make Normalization using MinMaxScaler
"""""
del Encoded_df['Name']
del Encoded_df['player_club']
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()

column_names = Encoded_df.columns
scaled_df = pd.DataFrame(scaler.fit_transform(Encoded_df), columns=column_names)
scaled_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.title("number of Wins VS number of scored goals")
sns.regplot(x=scaled_df['Wins'], y=scaled_df['player_goals']);
```

### This Scatter plot shows the correlation between the player's win matches and his goals
#### the main insight from that figure is:
- there is a positive correlation between player's win matches and his goals
- the more the player wins the more he scores goals
